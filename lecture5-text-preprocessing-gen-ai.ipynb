# Remove HTML tags

```python
import re
def remove_html_tags(text):
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)
```

# Remove URL

```python
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)
```

```python
data.loc[:,'review']=data['review'].apply(remove_url)
```

```python
import string,time
```

```python
exclude=string.punctuation
exclude
```

```python
def remove_punc(text):
    for char in exclude:
        text=text.replace(char,'')
    return text    
```

# Chat_conversion handle

```python
chat_words ={
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
    "EOM":"End of mail",
}
```

```python
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)        
```

```python
chat_conversion('EOM')
```

# stopwords

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
len(stopwords.words('english'))
```

```python
def remove_stopwords(text):
    new_text=[]
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append(' ')
        else:
            new_text.append(word)
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
            
```

```python
remove_stopwords('certain conditions during several generations are modified in the same manner')
```

```python
data.loc[:,'review']=data['review'].apply(remove_stopwords)
```

# remove_Emoji

```python
import re
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
        "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
        "\U0001F1E0-\U0001F1FF"  # Flags (iOS)
        "\U00002700-\U000027BF"  # Dingbats
        "\U000024C2-\U0001F251"  # Enclosed characters
        "]+", 
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
    
```

```python
data.loc[:,'review']=data['review'].apply(remove_emojis)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DistilBertTokenizer, DistilBertForSequenceClassification
```

```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
result=classifier("I am not a good person")
print(result)
```

```python
model1="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
model1_token=DistilBertTokenizer.from_pretrained(model1)
model1_classification=DistilBertForSequenceClassification.from_pretrained(model1)
```

```python
model2="nlptown/bert-base-multilingual-uncased-sentiment"
model2_token=AutoTokenizer.from_pretrained(model2)
model2_classification=AutoModelForSequenceClassification.from_pretrained(model2)

classifier=pipeline("sentiment-analysis",model=model2_classification,tokenizer=model2_token)
result=classifier("I am not a good person")
print(result)
```

```python
from transformers import AutoTokenizer

tokenizer=AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")

text="I am good person"

tokens=tokenizer.tokenize(text)
print(tokens)
```

```python
# Convert tokens to input IDs

input_ids=tokenizer.convert_tokens_to_ids(tokens)
print(input_ids)
```

```python
# Encode the text (tokenization + converting  to input IDs)

encoded_input=tokenizer(text)
print(encoded_input)   # 102 is a start token and 102 is a end token
```

```python
# Decode the text

decoded_output=tokenizer.decode(encoded_input['input_ids'])
print(decoded_output)
```

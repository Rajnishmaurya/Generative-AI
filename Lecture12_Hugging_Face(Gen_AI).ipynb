# NLP Tasks

Sentiment Analysis

```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
result=classifier("I am so not happy with the last Mission Impossible Movie")
print(result)
```

```python
pipeline(task="sentiment-analysis")("I was cofused with the Barbie Movie")
```

```python
# Multi line

pipeline(task="sentiment-analysis")\
                                    ("Everyday lots of LLMs papers are published about LLMs evalucation.\
                                     Lots of them looks very promising.\
                                     I am not sure if we cal actually evaluate LLMs.\
                                     There is still lots to do.\
                                     Don't you think? ")
```

```python
pipeline(task="sentiment-analysis",model="BAAI/bge-reranker-v2-m3")\
                                                                   ("Everyday lots of LLMs papers are published about LLMs evalucation.\
                                                                    Lots of them looks very promising.\
                                                                    I am not sure if we cal actually evaluate LLMs.\
                                                                    There is still lots to do.\
                                                                     Don't you think? ")
```

Batch sentiment Analysis

```python
classifier=pipeline("sentiment-analysis")
task_list=["Everyday lots of LLMs papers are published about LLMs evalucation.",
           "Lots of them looks very promising.",
           "I am not sure if we cal actually evaluate LLMs.",
           "There is still lots to do.",
          "Don't you think?"]
classifier(task_list)
```

```python

classifier=pipeline("sentiment-analysis",model="SamLowe/roberta-base-go_emotions")
task_list=["Everyday lots of LLMs papers are published about LLMs evalucation.",
           "Lots of them looks very promising.",
           "I am not sure if we cal actually evaluate LLMs.",
           "There is still lots to do.",
          "Don't you think?"]
classifier(task_list)
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comments",
                         "dswithbappy write comments"],"outputs":[1,1,0,0]})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
bag_of_words=cv.fit_transform(df['text'])
```

```python
bag_of_words
```

```python
print(cv.vocabulary_)
```

# N-grams

```python
df=pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comments",
                         "dswithbappy write comments"],"outputs":[1,1,0,0]})
```

```python
# bi-grams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(2,2))
```

```python
bow=cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
bow.toarray()
```
